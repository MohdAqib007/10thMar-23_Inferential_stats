```python
#Q1: What is Estimation Statistics? Explain point estimate and interval estimate.
#Ans-
'''Estimation statistics is a branch of statistics that deals with making predictions or inferences about a population based on sample data. 
It involves using statistical methods to estimate population parameters such as means, variances, and proportions.

Point Estimate:
A point estimate is a single value that is used to estimate an unknown population parameter. 
It is obtained by calculating a statistic, such as the sample mean or sample proportion, from a sample of data. 
For example, the sample mean of a group of 50 studentsâ€™ test scores can be used as a point estimate of the population mean test score.

Interval Estimate:
An interval estimate is a range of values that is used to estimate an unknown population parameter. 
It is obtained by calculating a confidence interval, which is a range of values that is likely to contain the true population parameter with a certain level of confidence. 
For example, a 95% confidence interval for the population mean test score would be a range of values calculated from the sample data that is likely to contain the true mean test score with a 95% level of confidence.'''
```

```python
#Q2. Write a Python function to estimate the population mean using a sample mean and standard deviation.
#Ans-
'''Sure, here's an example Python function that estimates the population mean using a sample mean and standard deviation:'''

import math

def estimate_population_mean(sample_mean, sample_std_dev, sample_size):
    """Estimates the population mean using a sample mean and standard deviation"""
    # Calculate the standard error of the mean
    std_err = sample_std_dev / math.sqrt(sample_size)
    # Calculate the lower and upper bounds of the 95% confidence interval
    lower_bound = sample_mean - 1.96 * std_err
    upper_bound = sample_mean + 1.96 * std_err
    # Return the estimated population mean and confidence interval
    return sample_mean, (lower_bound, upper_bound)
```

```python
'''To use this function, you would need to provide it with the sample mean, sample standard deviation, and sample size as arguments. 
The function then calculates the standard error of the mean, which is the standard deviation of the sampling distribution of the mean. 
It then calculates the lower and upper bounds of the 95% confidence interval using the formula sample_mean +/- 1.96 * std_err, where 1.96 is the z-score corresponding to a 95% confidence interval.

Here's an example usage of the function:'''

# Example usage
sample_mean = 10.5
sample_std_dev = 2.3
sample_size = 100
population_mean, confidence_interval = estimate_population_mean(sample_mean, sample_std_dev, sample_size)
print(f"Estimated population mean: {population_mean:.2f}")
print(f"95% Confidence interval: {confidence_interval}")

#This means that we estimate the population mean to be around 10.5, with a 95% confidence interval ranging from 10.01 to 10.99.
```

```python
#Q5. Write a Python script to conduct a hypothesis test on the difference between two population means, given a sample from each population.
#Ans-
'''Here's an example Python script to conduct a hypothesis test on the difference between two population means, given a sample from each population:'''

import numpy as np
from scipy import stats

# Generate two random samples from normal distributions with different means and variances
sample1 = np.random.normal(loc=10, scale=2, size=50)
sample2 = np.random.normal(loc=12, scale=3, size=50)

# Conduct a two-sample t-test to test if the difference between the sample means is statistically significant
t_stat, p_val = stats.ttest_ind(sample1, sample2)

# Print the test results
print("Test results:")
print(f"t-statistic: {t_stat:.3f}")
print(f"p-value: {p_val:.3f}")

# Interpret the results
alpha = 0.05
if p_val < alpha:
    print(f"The p-value of {p_val:.3f} is less than the significance level of {alpha}.")
    print("We reject the null hypothesis that the two population means are equal.")
    print("There is strong evidence to suggest that the population means are different.")
else:
    print(f"The p-value of {p_val:.3f} is greater than the significance level of {alpha}.")
    print("We fail to reject the null hypothesis that the two population means are equal.")
    print("There is no strong evidence to suggest that the population means are different.")
```

```python
'''In this script, we first generate two random samples from normal distributions with different means and variances using the NumPy library. 
We then use the ttest_ind function from the SciPy library to conduct a two-sample t-test, which tests if the difference between the sample means is statistically significant.

We then print the test results, including the t-statistic and p-value. Finally, we interpret the results by comparing the p-value to a significance level (alpha) of 0.05. 
If the p-value is less than alpha, we reject the null hypothesis that the two population means are equal and conclude that there is strong evidence to suggest that the population means are different. 
If the p-value is greater than alpha, we fail to reject the null hypothesis and conclude that there is no strong evidence to suggest that the population means are different.'''
```

```python
#Q9. Generate a Student's t-distribution plot using Python's matplotlib library, with the degrees of freedom parameter set to 10.
#Ans-
'''Here's an example Python code to generate a Student's t-distribution plot with 10 degrees of freedom using the matplotlib library:'''

import numpy as np
import matplotlib.pyplot as plt

# Set the degrees of freedom
df = 10

# Define the range of x values for the t-distribution
x = np.linspace(-4, 4, 1000)

# Calculate the corresponding y values for the t-distribution
y = np.exp(-x ** 2 / 2) / np.sqrt(2 * np.pi) / np.sqrt(df) / np.power(1 + x ** 2 / df, (df + 1) / 2)

# Create the plot
plt.plot(x, y)
plt.title("Student's t-distribution (df=10)")
plt.xlabel("x")
plt.ylabel("Probability density")
plt.show()
```

```python
#Q10. Write a Python program to calculate the two-sample t-test for independent samples, given two random samples of equal size and a null hypothesis that the population means are equal.
#Ans-
'''This code generates two random samples using the normal function from numpy, assuming that the population means are 10 and 12, and the standard deviations are 2 for both populations. 
It then calculates the means and standard deviations of the two samples, and uses them to calculate the pooled standard deviation. 
It then calculates the t-statistic using the formula (mean1 - mean2) / (pooled_std * np.sqrt(2 / len(sample1))). 
The degrees of freedom are calculated as df = len(sample1) + len(sample2) - 2. Finally, the p-value is calculated using the t.cdf function from the stats module of scipy. The results are printed to the console.

Note that in this example, we assume that the variances of the two populations are equal, which is why we use the pooled standard deviation. 
If the variances are not equal, we need to use a modified version of the t-test called the Welch's t-test, which does not assume equal variances.'''

import numpy as np
from scipy import stats

# Generate two random samples with equal size
sample1 = np.random.normal(loc=10, scale=2, size=50)
sample2 = np.random.normal(loc=12, scale=2, size=50)

# Calculate the means and standard deviations of the two samples
mean1 = np.mean(sample1)
mean2 = np.mean(sample2)
std1 = np.std(sample1, ddof=1)
std2 = np.std(sample2, ddof=1)

# Calculate the pooled standard deviation
pooled_std = np.sqrt(((len(sample1) - 1) * std1 ** 2 + (len(sample2) - 1) * std2 ** 2) / (len(sample1) + len(sample2) - 2))

# Calculate the t-statistic
t_statistic = (mean1 - mean2) / (pooled_std * np.sqrt(2 / len(sample1)))

# Calculate the degrees of freedom
df = len(sample1) + len(sample2) - 2

# Calculate the p-value
p_value = 2 * stats.t.cdf(-np.abs(t_statistic), df)

# Print the results
print("Sample 1 mean: {:.2f}".format(mean1))
print("Sample 2 mean: {:.2f}".format(mean2))
print("Pooled standard deviation: {:.2f}".format(pooled_std))
print("t-statistic: {:.2f}".format(t_statistic))
print("p-value: {:.4f}".format(p_value))
```
